Count the number of orders placed in 2017.

```python
query="""select count(order_id) as Total_orders_2017 from orders 
where  year(order_purchase_timestamp)=2017;"""
pd.read_sql(query,conn)
```

Find the total sales per category.

```python
query="""SELECT 
    upper(products.productcategory) category, round(SUM(payments.payment_value),2) as sales
FROM
    products
        JOIN
    order_items ON products.product_id = order_items.product_id
        JOIN
    payments ON payments.order_id=order_items.order_id
GROUP BY category order by sales desc"""
pd.read_sql(query,conn)
```

calculate the percentage of orders that were paid in installments(EMI).

```python
query="""select (sum(case when payment_installments>=1 then 1 else 0 end)/count(*)*100) as paid_installment_percentage
from payments;"""
pd.read_sql(query,conn)
```

Count number of customers from each state.

```python
query="""select customer_state,count(customer_id) as customer_count from customers 
group by customer_state order by customer_count desc"""
df=pd.read_sql(query,conn)
df
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["customer_state"],df["customer_count"])
plt.title("Number of customer of each state")
plt.xlabel("State")
plt.ylabel("Customer_count")
plt.show()
```

Calculate the number of orders per month in 2018.

```python
query="""select monthname(order_purchase_timestamp) as Month,count(order_id) as order_count from orders
where year(order_purchase_timestamp)=2018 group by Month ;"""
df=pd.read_sql(query,conn)
df
```

```python
plt.figure(figsize=(10,5))
ax=sns.barplot(x="Month",y="order_count",data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of order by months 2018")
plt.show()
```

Calculate number of order based on hours 2018.

```python
query="""select hour(order_purchase_timestamp) as Hours, count(order_id) as order_count
from orders
where year(order_purchase_timestamp)=2018 group by Hours order by Hours;"""
df=pd.read_sql(query,conn)
df
```

```python
plt.figure(figsize=(11,5))
sns.lineplot(x=df["Hours"],y=df["order_count"],data=df,marker="o")
plt.title('Order based on Hours')

plt.show()
```

Find the Average number of product per order.

```python
query="""select orders.order_id,avg(order_items.order_item_id)as Avg_product_per_order from orders
join order_items on orders.order_id=order_items.order_id group by orders.order_id order by Avg_product_per_order desc"""

df=pd.read_sql(query,conn)
df.head(10)
```

Calculate the percentage of total revenue contributed by each product category.

```python
query="""select products.productcategory category,round((sum(payments.payment_value)/
(select sum(payment_value) from payments))*100,2) 
as Sales_percentage from products 
join 
order_items on products.product_id=order_items.product_id
join
payments on payments.order_id=order_items.order_id
group by category order by Sales_percentage desc"""

df=pd.read_sql(query,conn)
df.head(10)

```

```python
plt.figure(figsize=(9,5))
ax=sns.barplot(x=df["category"].head(),y=df['Sales_percentage'].head())
ax.bar_label(ax.containers[0])
plt.title("Percentage of revenue of top 5 product category")
plt.xticks(rotation=30)
plt.show()
```

Find the numbers of time product purchased and Avg price.

```python
query=""" select upper(products.productcategory) as product_category,count(order_items.product_id)as purchase_product_count,avg(order_items.price)
as Avg_price from products
join 
order_items on products.product_id=order_items.product_id 
group by product_category order by purchase_product_count desc""" 
df=pd.read_sql(query,conn)
df
```

```python
plt.figure(figsize=(9,5))
ax=sns.barplot(x=df["product_category"].head(),y=df['purchase_product_count'].head())
ax.bar_label(ax.containers[0])
plt.title("Top 5 product category Based on purchase_count")
plt.xticks(rotation=25)
plt.show()
```

Calculate the total revenue genarated by each seller and rank than by revenue.

```python
query="""select *,rank() over(order by revenue desc) 
as Revenue_rank from(select order_items.seller_id,sum(payments.payment_value)
as Revenue from order_items 
join 
payments on order_items.order_id=payments.order_id 
group by order_items.seller_id)as a """ 
df=pd.read_sql(query,conn)
df.head()
```

calculate the cumulative sales per month for each year.

```python
query="""select years,months,sales,sum(sales) over(order by years,months) as cumulative_sales from
(select year(orders.order_purchase_timestamp) as years,month(orders.order_purchase_timestamp) as months,
round(sum(payments.payment_value),2) as sales from orders 
join
payments on orders.order_id=payments.order_id 
group by years,months order by years,months)as a;"""
df=pd.read_sql(query,conn)
df
```

Calculate year over year growth rate of total sales.

```python
query="""with a as
(select year(orders.order_purchase_timestamp) as years,round(sum(payments.payment_value),2) as sales from orders 
join 
payments on orders.order_id=payments.order_id 
group by years)
select years,((sales-lag(sales,1) over(order by years))/ lag(sales,1) over(order by years))*100 as Sales_Growth from a"""
pd.read_sql(query,conn)
```

Identify Top 3 customers who spend the most money in each year.

```python
query="""select years,customer_id,payment from
(select year(orders.order_purchase_timestamp)as years,orders.customer_id,sum(payments.payment_value) as payment,
dense_rank()over(partition by year(orders.order_purchase_timestamp) order by sum(payments.payment_value) desc)as d_rank from orders 
join 
payments on orders.order_id=payments.order_id 
group by orders.customer_id,years) as a 
where d_rank<4;"""

df=pd.read_sql(query,conn)
df
```

```python
sns.barplot(x="customer_id",y="payment",hue="years",data=df)
plt.xticks(rotation=90)
plt.title("Top 3 customers spend the most money in each year")
plt.show()
```

Cities with the Most Orders from Sellers.

```python
query="""select sellers.seller_city,count(orders.order_id) as orders_count from sellers
join 
order_items on sellers.seller_id=order_items.seller_id 
join 
orders on orders.order_id=order_items.order_id 
group by sellers.seller_city 
order by orders_count desc ;"""
df=pd.read_sql(query,conn)
df
```

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(x=df["seller_city"].head(),y=df["orders_count"].head())
ax.bar_label(ax.containers[0])
plt.title('Top Cities Based on Seller Orders')
plt.show()
```
